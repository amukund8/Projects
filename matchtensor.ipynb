```python
import torch
import torch.nn.functional as F

# Create a dummy tensor of size [64, 2]
dummy_tensor = torch.randn(64, 2)
soft= F.softmax(dummy_tensor, dim=1)

# Find the index of the maximum value along a specific dimension (e.g., dimension 1)
max_indices = torch.argmax(soft, dim=1)

print("Dummy Tensor:")
print(dummy_tensor)

print("Softmax of Dummy Tensor:")
print(soft)
print(soft.size())

print("\nIndices of Maximum Values along Dimension 1:")
print(max_indices)

print("\nTensor Size of Maximum Indices:")
print(max_indices.size())

```
